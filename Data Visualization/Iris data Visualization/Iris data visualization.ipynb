```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r'D:\Shrikant\Python and data science\IRIS DATASET _ ADVANCE VISUALIZATION _ EDA 2\iris.csv')
```

```python
data.sample(5)
```

```python
data.info()
```

```python
data.Species.value_counts()
```

```python
sns.countplot(data=data,x='Species')
```

```python
sns.jointplot(data=data,x='SepalLengthCm',y='SepalWidthCm',hue='Species')
```

```python
sns.jointplot(data=data,x='SepalLengthCm',y='SepalWidthCm',kind='hex')
```

```python
sns.FacetGrid(data,hue='Species')\
.map(plt.scatter,'SepalLengthCm','SepalWidthCm')\
.add_legend()
```

```python
fig = plt.gcf()
fig.set_size_inches(10,7)
sns.boxplot(data=data,x='Species',y='PetalLengthCm',order=['Iris-setosa','Iris-versicolor','Iris-virginica'],linewidth=2.5,dodge=False)
```

```python
data.boxplot(by='Species',figsize=(12,6))
```

```python
sns.pairplot(data=data,kind='scatter',hue='Species')
```

```python
sns.pairplot(data=data,kind='scatter')
```

```python
fig = plt.gcf()
fig.set_size_inches(10,7)
sns.heatmap(data.corr(),annot=True,cmap='cubehelix',linewidth=1,linecolor='k',square=True,mask=False,vmin=-1,vmax=1,cbar_kws={'orientation':'vertical'},cbar=True)
```

```python
data.hist(edgecolor='black',linewidth=1.2)
fig = plt.gcf()
fig.set_size_inches(12,6)
```

```python
sns.set_theme(style='darkgrid')
fig = plt.gcf()
fig.set_size_inches(10,7)
sns.swarmplot(x='Species',y='PetalLengthCm',data=data)
```

```python
data.plot.area(y=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],alpha=0.4,figsize=(12,6))
```
